```
import numpy as np

# Create a NumPy array
arr = np.array([45, 12, 89, 33, 5, 72])

print("Original Array:", arr)

# Sort the array
sorted_arr = np.sort(arr)

print("Sorted Array:", sorted_arr)
```

```
sorted_desc = np.sort(arr)[::-1]
print("Descending Order:", sorted_desc)
```
